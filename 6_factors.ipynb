```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

file = 'data1.csv'
data = pd.read_csv(file)
data.columns = ['weekly_rents_per_bedroom', 'latitude', 'longitude', 'distance_to_CBD', 'bedrooms','market','underground']
x = sm.add_constant(data.iloc[:,1:]) #生成自变量
y = data['weekly_rents_per_bedroom'] #生成因变量
model = sm.OLS(y, x) #生成模型
result = model.fit() #模型拟合
result.summary() #模型描述
```
